### Btech CGPA

```python
def to_num_cgpa(x):
    '''
    function to parse 'BTech CGPA' into numerical values
    '''
    try:
        return float(x)
    except:
        return x.split('(')[0]
```

```python
data_new['btech cgpa'] = pd.to_numeric(data_new['btech cgpa'].apply(to_num_cgpa))
```

```python
def preprocessor(data: dict)-> dict:
    
    # values for validation
    valid_dict = {
        'age': {'min': 19, 'max': 24, 'mean': 21},
        '10th percentage': {'min': 49.5, 'max': 95.0, 'mean': 87.12},
        '12th/diploma percentage': {'min': 50.0, 'max': 94.0, 'mean': 77.85},
        'btech cgpa': {'min': 6.35, 'max': 8.95, 'mean': 7.89}
    }
    
    # column maps
    label_map = {'gender': {'Female': 0, 'Male': 1},
     'dropper?': {'No': 0, 'Yes': 1},
     'preparation level for interview': {'High': 0, 'Low': 1, 'Moderate': 2},
     'internships done?': {'No': 0, 'Yes': 1},
     'placed?': {'No': 0, 'Yes': 1}}
    
    
    def assert_type(col, data_types):
        
        count_types = 0
        
        if type(data_types)==list:
            for data_type in data_types:
                if type(data[col])==data_type:
                    pass
                else:
                    count_types+=1
                    
        else:
            if type(data[col])==data_types:
                pass
            else:
                count_types+=1
        
        if count_types==0:
            raise TypeError(col+" needs to be of atleast type: "+ str(data_types))
    
    def fix_missing_or_random(col, min_val, max_val, fill_val):
        if data[col]<min_val or data[col]>max_val:
            data[col] = fill_val
            
    # Age  
    assert_type('age', int)
    fix_missing_or_random(col, valid_dict['age']['min'], valid_dict['age']['max'], valid_dict['age']['mean'])
        
    # 10th percentage
    assert_types('10th percentage', [int, float])
    fix_missing_or_random(col,
                          valid_dict['10th percentage']['min'],
                          valid_dict['10th percentage']['max'],
                          valid_dict['10th percentage']['mean'])
        
    # 12th/Diploma percentage
    assert_types('12th/diploma percentage', [int, float])
    fix_missing_or_random(col,
                          valid_dict['12th/diploma percentage']['min'],
                          valid_dict['12th/diploma percentage']['max'],
                          valid_dict['12th/diploma percentage']['mean'])
    
    # Btech CGPA
    assert_types('12th/diploma percentage', [int, float])
    fix_missing_or_random(col,
                          valid_dict['12th/diploma percentage']['min'],
                          valid_dict['12th/diploma percentage']['max'],
                          valid_dict['12th/diploma percentage']['mean'])
    # Gender
    assert_types('gender', str)
    if data['gender'] not in label_map['gender']:
        raise ValueError('Gender needs to either:'+ str(list(label_map['gender'].keys())))
    
    data['gender'] = label_map['gender'][data['gender']]
    
    # dropper?
    assert_types('dropper?', str)
    if data['dropper?'] not in label_map['dropper?']:
        raise ValueError('dropper needs to either:'+ str(list(label_map['dropper?'].keys())))
    
    data['dropper?'] = label_map['dropper?'][data['dropper?']]
    
    # preparation level for interview
    assert_types('preparation level for interview', str)
    if data['preparation level for interview'] not in label_map['preparation level for interview']:
        raise ValueError('preparation level for interview:'+ str(list(label_map['preparation level for interview'].keys())))
    
    data['preparation level for interview'] = label_map['preparation level for interview'][data['preparation level for interview']]
    
    # internships done?
    assert_types('internships done?', str)
    if data['internships done?'] not in label_map['internships done?']:
        raise ValueError('internships done? needs to be either:'+ str(list(label_map['internships done?'].keys())))
    
    data['internships done?'] = label_map['internships done?'][data['internships done?']]
    
    # placed?
    assert_types('placed?', str)
    if data['placed?'] not in label_map['placed?']:
        raise ValueError('placed? needs to be either:'+ str(list(label_map['placed?'].keys())))
    
    data['placed?'] = label_map['placed?'][data['placed?']]
    
    # board
    assert_types('board', str)
    board_valid = ['CBSE', 'CHSE', 'ICSE']
    if data['board']=='':
        data['board'] = 'CBSE'
        
    if data['board'] not in board_valid:
        raise ValueError('board needs to be either:'+ str(board_valid))
        
    # hobby
    assert_types('hobby', str)
    board_valid = ['Art', 'Coding', 'Dance', 'Music', 'Others', 'Sports', 'nothing_particular']
    if data['hobby']=='':
        data['hobby'] = 'nothing_particular'
        
    if data['hobby'] not in board_valid:
        raise ValueError('hobby needs to be either:'+ str(board_valid))
        
    # branch
    assert_types('branch', str)
    board_valid = ['CSE', 'Mechanical', 'Mettalurgy', 'ETC', 'Electrical', 'Civil',
       'Production', 'Chemical']
    if data['branch']=='':
        data['branch'] = 'nothing_particular'
        
    if data['branch'] not in board_valid:
        raise ValueError('branch needs to be either:'+ str(board_valid))
        
    # interest
    assert_types('interest', str)
    board_valid = ['Banking', 'MBA', 'MS', 'MTech']
    if data['interest'] not in board_valid:
        data['interest'] = 'Others'
        
    if data['interest'] not in board_valid:
        raise ValueError('interest needs to be either:'+ str(board_valid))
```
